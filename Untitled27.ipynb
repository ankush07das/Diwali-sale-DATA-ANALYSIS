Age Analysis

```python
ax =sns.countplot(x='Age Group',data=df, hue='Gender')
for bars in ax.containers:
  ax.bar_label(bars)
  plt.title('Age Group vs Gender')
  plt.ylabel("Number of people")
```

From the above graph we conclude that most of the buyers are in the age group of 26-35 and mostly females

```python
#Total amount vs Age group
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=True)
sns.barplot(x='Age Group',y='Amount',data=sales_age,hue='Age Group')
plt.title('Total Amount vs Age Group)')
```

Analysing the States now

```python
states_data=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=True)
sns.barplot(x='State',y='Orders',data=states_data,hue='Orders')
sns.set(rc={'figure.figsize':(30,5)})
plt.title('State vs Orders')
```

From here we can conclude orders are maximum from Madhya Pradesh

Analysing the Occupation of the buyers

```python
sns.set(rc={'figure.figsize':(25,2)})
ax=sns.countplot(x='Occupation',data=df,hue='Gender')
for bars in ax.containers:
  ax.bar_label(bars)
plt.title('Occupation vs Gender')
plt.xlabel('Occupation')
plt.ylabel('Number of people')
```

```python
sns.set(rc={'figure.figsize':(45,5)})
ax=sns.countplot(x='Occupation',data=df,hue='Age Group')
for bars in ax.containers:
  ax.bar_label(bars)
plt.title('Occupation vs Age Group')
plt.xlabel('Occupation')
plt.ylabel('Number of people')
```

From the above we can conclude that maximum Customers belong to the IT Sector

Analysing Product Selling

```python
sns.set(rc={'figure.figsize':(30,5)})
ax=sns.countplot(x='Product_Category',data=df,hue='Gender')
for bars in ax.containers:
  ax.bar_label(bars)
  plt.title('Product Category vs Gender')
  plt.xlabel('Product Category')
  plt.ylabel('Number of people')
```

```python
sales_state_data=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(35,5)})
sns.barplot(x='Product_Category',y='Amount',data=sales_state_data,hue='Product_Category')
plt.title('Product Category vs Amount')
plt.xlabel('Product Category')
plt.ylabel('Amount')
```

```python
sales_state_data=df.groupby(['Product_Category'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(35,5)})
sns.barplot(x='Product_Category',y='Orders',data=sales_state_data,hue='Product_Category')
plt.title('Product Category vs Orders')
plt.xlabel('Product Category')
plt.ylabel('Orders')
```

From the above graphs we get to know that Food and clothing are ordered maximum times

Conclusion : Married womans in the group of 26-35 from the states of Uttar Pradesh , Madhyapradesh and Karnataka from the It sector and the health care sector ordered maximum times on clothing, food and electronics items.
